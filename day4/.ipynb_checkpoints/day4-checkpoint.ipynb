```python
# Tht ton khng dng remove() v count()
n=int(input()) # S lng danh sch
a=[] # Lu danh sch
for i in range(n):
    x=int(input())
    a.append(x) # a s nguyn vo list a

a.sort()
print(a)

b=[a[0]] # Khi to list b c gi tr a[0]


for i in range (1, len(a)): # i = 0
    if (a[i]!=a[i-1]):
        b.append(a[i])
print(b)
```

```python
lst = sorted([int(x) for x in input().split()])

n = int(input())
lst1 = []
lst2 = []
for i in range (len(lst)):
    if lst[i] < n :
        lst1.append(lst[i])
    else:
        lst2.append(lst[i])
print(*lst1)
print(*lst2)
```
