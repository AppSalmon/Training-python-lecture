```python
# Thật toán không dùng remove() và count()
n=int(input()) # Số lượng danh sách
a=[] # Lưu danh sách
for i in range(n):
    x=int(input())
    a.append(x) # Đưa số nguyên vào list a

a.sort()
print(a)

b=[a[0]] # Khởi tạo list b có giá trị a[0]


for i in range (1, len(a)): # i = 0
    if (a[i]!=a[i-1]):
        b.append(a[i])
print(b)
```

```python
lst = sorted([int(x) for x in input().split()])

n = int(input())
lst1 = []
lst2 = []
for i in range (len(lst)):
    if lst[i] < n :
        lst1.append(lst[i])
    else:
        lst2.append(lst[i])
print(*lst1)
print(*lst2)
```
